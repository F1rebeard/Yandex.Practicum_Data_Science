## Предобработка данных

Посмотрим в каких столбцах и в каком количестве имеются пропуски.

```python
#применяем метод isna() и смотрим сумму пропусков по столбцам.
data.isna().sum()
```

### Выявленные пропуски

1. Рассмторим выявленные пропуски:
2. Пропуски в столбце с высотой потолка. Возможная причина - человек не смог замерить высоту или не знает её, соотвественно не указал её.
3. Пропуски с количеством этажей в доме. Возможная причина - забыли указать (человеский фактор) или ошибка в данных. Таких пропусков всего 86.
4. Пропуски с жил.площадью. Такие данные являются важными и пропустить заполнение таких данных - вызывает подозрение.
5. Пропуски с аппартаментами. Возможно данный вид категории вызвал замешательство у людей. Надо посмотреть на другие значения этого столбца. Возможно так из заполненных только логическое "Нет".
6. Пропуски с площадью кухни. Такие данные являются важными и пропустить заполнение таких данных - вызывает подозрение.
7. Пропуски с количеством балконов. Возможно просто отуствие балконов в данной квартире.
8. Не заполнено название населенного пункта, всего 49 пропусков. Похоже на человеческую невнимательность или же это новостройка в середине поля.
9. Пропуски с расстоянием до ближайшего аэропорта. Возможная причина пропуска - человек не знает. Аналогично с расстоянием до центра города, количеством парков в радиусе 3 км и водоемами.
10. Пропуски в днях размещения объявления. Возможная причина - то что объявление ещё осталось. Поэтому и количество дней до снятия не указано.

Также причины возможного пропуска - неполностью заполненные объявления от людей.

### Пропуски в столбце "Количество балконов"

Рассммотрим пропуски с количеством балконов. Возможный человеческий фактор - отсуствие балконов оставили пустым.
Соотвественно заменяем все строчки с пропущенным значением балкона на "0".

```python
#применяем метод fillna() к пустым значениям в столбце "balcony" и проверяем
data['balcony'] = data['balcony'].fillna(0)
data['balcony'].isna().sum()
```

### Пропуски в столбце "Аппартаменты"

Пропуски в столбце с определнием аппартаментов считаю тоже человеским фактором, и пропуск был сделан в случае того, что помещения является жилым и относится к жилому фонду. Заменим пустые значение на логическое отрицание.

```python
#применяем метод fillna() к пустым значениям в столбце "is_apartment" и проверяем
data['is_apartment'] = data['is_apartment'].fillna(False)
data['is_apartment'].isna().sum()
```

### Пропуски в количестве дней от публикации до снятия объявления

Пропуски в столбце с количеством дней от публикации до снятия выявлен тем, что на момент получения датафрейма эти объявления ещё "висят" и квартира не продана. Предлагаю заменить эти пропущенные значения на "0". То есть "0" в столбце с днями означает, что объявление ещё активно.

### Расчет цены квадратного метра

Рассчитаем цену квадратного метра в новом столбце таблицы, для этого поделим стоимость квартиры на общую площадь.

### Добавление категоризации по этажу. Первый, последний, другой.

Добавим новый столбец в таблицу с параметрами как в старом вестерне с Клинтом Иствудом "Хороший, плохой, злой" - Первый, последний, другой. Для этого сделаем функцию которая создаст нам такие категории по этажам в столбце нашего датафрейма.

```python
# функция для сортировки этажей:

def floor_groups(data):
    """
    Возвращает категорию этажа квартиры со следующими правилами:
    - этаж = 1 - "Первый"
    - 1 < этаж < последний - "Другой"
    - этаж = последний - "Последний"
    
    """
        
    floor = data['floor']
    floors_total = data['floors_total']
    if floor == 1:
        return 'Первый'
    elif 1 < floor < floors_total:
        return 'Другой'
    elif floor == floors_total:
        return 'Последний'

# применим функцию к датафрейму, результат сохраним в новом столбце 'floor_category'
data['floor_groups'] = data.apply(floor_groups, axis=1)


data.head(10)
```

### Добавление соотношений  площадей жилая/общая и кухня/общая.

Добавим два новых столбца с соотношением жилой площади к общей, и с площадью кухни к общей.

```python
# добавим соотношения в новых столбцах 'kitchen/total_ratio' и 'living/total_ratio'

data['kitchen/total_ratio'] = data['kitchen_area'] / data['total_area']
data['living/total_ratio'] = data['living_area'] / data['total_area']

#округлим до 2 знаков после запятой
data['kitchen/total_ratio'] = data['kitchen/total_ratio'].round(decimals=2)
data['living/total_ratio'] = data['living/total_ratio'].round(decimals=2)

data.head(10)
```

### Изучение факторов влияния на стоимость квартир

Рассмотрим факторы влияния на стоимость квартир, для этого оценим корреляцию по коэффициенту Пирсона стоимости от других параметров: квадратного метра, число комнат, этажа, удаленности от центра, и даты размещения.

#### Стоимость квартиры от  квадратного метра, числа комнат и этажа

```python
#создадим функцию для создания численных индексов сортировки квартиры по этажам
def floor_groups_id(floor):
    """
    Возвращает категорию этажа квартиры со следующими правилами:
    - этаж "Первый" - floor_id = 1
    - этаж "Другой" - _floor_id = 2
    - этаж "Последний" - _floor_id = 3
    
    """
          
    if floor == 'Первый':
        return 1
    elif floor == 'Другой':
        return 2
    elif floor == 'Последний':
        return 3

# применим функцию к cтолбцу с категориями этажей, результат сохраним в новом столбце 'floor_id'
data['floor_id'] = data['floor_groups'].apply(floor_groups_id)
```

Сначала посмотрим корреляцию цены от параметров квартиры - площадь, количество комнат и этаж. Создадим матрицу диаграмм рассеяния для этих параметров, и оценим корреляцию этих параметров.

```python
total_price_flat_char = data[['last_price_mln','total_area', 'rooms', 'floor_id']]

pd.plotting.scatter_matrix(total_price_flat_char, figsize = (14, 14))
```

На графике цены от количество комнат, можно наблюдать нормальное распределение пиков в районе 4 комнат, и небольшие точечные выбросы по цене квартиры.

На графике цеты от этажа, где 1 - первый, 2 - другой, и 3 - последний этаж. Видно, что количество и высшая стоимость у последних и не первых этажей. Квартир с первым этажом меньше, и ценовой разброс меньше и ниже.

Оценим корреляцию этих параметров.

```python
#посмотрим корреляцию по Пирсону
total_price_flat_char.corr().round(decimals=2)
```

Из табилцы видно, что корреляция цены с количеством комнат составляет - 0.36. А с категорией этажа - 0.06. Количество комнат и цена имеют слабую корреляцию, а корреляция у категории этажа и стоимость - практически отсусвует.

### Топ-10 населенных пунктов

Выделим топ-10 населенны пунктов и вычислим среднюю цену одного м2 для каждого из этих пунктов. Отсортируем по цене квадратного метра и найдем максимальную и минимальную цену по населенным пунктам.
